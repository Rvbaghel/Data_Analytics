```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df=pd.read_csv('emp.csv')

df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
x=df.groupby('Position')['Salary'].mean()

x=x.sort_values()
plt.figure(figsize=(10,5))

plt.title('Average Salary of each Position')
plt.barh(x.index,x.values)
plt.xlabel("Salary Avarage")
plt.show()
```

```python
fre=df['Gender'].value_counts() 
plt.title('Gender Distribution')
plt.pie(fre,labels=fre.index,autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.show()
```

```python
fre=df['Gender'].value_counts() 

plt.title('Gender Distribution')

plt.bar(fre.index,fre.values,color=['#bbbfff','#fffbbb'],width=0.5)
plt.xlabel('Genders')
plt.ylabel('Frequency')
plt.show()
```

```python

expe = df[(df['Experience (Years)'] >= 10) & (df['Experience (Years)'] <= 15)]

range_salay = expe['Salary']

plt.hist(range_salay, bins=5, edgecolor='black')
plt.title('Salary Distribution for 10-15 Years Experience')
plt.xlabel('Salary ($)')
plt.ylabel('Frequency')
plt.show()
```

```python

position_counts = df['Position'].value_counts()

plt.barh(position_counts.index, position_counts.values)

plt.title('Number of Employees by Position')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.grid(axis='y')
plt.show()
```

```python

plt.barh(df['Position'],df['Salary'])

plt.title('Number of Employees by Position')
plt.xlabel('Salary')
plt.ylabel('Position')
plt.grid(True)
plt.show()
```
